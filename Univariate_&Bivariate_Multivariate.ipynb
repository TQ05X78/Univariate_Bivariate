# Bivariate and Multivariate Analysis

# 1. ScatterPlot (Numerical to Numerical)

```python
sns.scatterplot(df['Fare'],df['Pclass'],hue=df['Age'],style=df['Survived'])
```

```python
sns.clustermap(pd.crosstab(df['Parch'],df['Survived']))
```
